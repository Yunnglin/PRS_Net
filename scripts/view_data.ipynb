```python
import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import nrrd
import random
import os
```

```python
# data_root = 'D:\\My\\MyCode\\PRSPaper\\PRSNet\\data\\bench'
data_root = 'D:\\My\\MyCode\\PRSPaper\\PRSNet\\data\\skateboard'
start_file = 1
file_count = 1  # 151
view = True
```

```python
for index in range(start_file, start_file+file_count):
    # read nrrd voxel data
#     print(index)
    nrrd_path = os.path.join(data_root, str(index), 'model.nrrd')
    try:
        print(index)
        nrrd_data, nrrd_head = nrrd.read(nrrd_path)
#         print(nrrd_head)
        transform = transforms.ToTensor()
        voxel = transform(nrrd_data)
        voxel = voxel.view(1, 32, 32, 32)
    
    
        # read pcd points data
        pcd_path = os.path.join(data_root, str(index), 'model.pcd')
        with open(pcd_path, mode='r') as pcd_data:
            # skip first 9 lines
            for i in range(9):
                pcd_data.readline()
            num_line = pcd_data.readline()
            num_points = num_line.split(' ')[1]
            pcd_data.readline()  # skip one more line
            points = []
            for i in range(int(num_points)):
                line = pcd_data.readline()
                point = line.split(' ')
                points.append([float(point[0]),
                               float(point[1]),
                               float(point[2])])
    #       print(num_points)

        point = torch.tensor(points)
        max, _ = torch.max(point, dim=0)
        min, _ = torch.min(point, dim=0)
        max_dis = torch.max(max)

        print(torch.norm(max+min))
    except Exception as e:
        pass

```

```python
data_root = '..\\data\\bench'
mark = '-skip'
for foldName, subfolders, filenames in os.walk(data_root):     #用os.walk方法取得path路径下的文件夹路径，子文件夹名，所有文件名
    for subfolder in subfolders:     #遍历列表下的子文件夹名
        if subfolder.find(mark) != -1:
            continue
        
#         # read nrrd voxel data
#         nrrd_path = os.path.join(data_root, subfolder, 'model.nrrd')
#         nrrd_data, nrrd_head = nrrd.read(nrrd_path)
#         # print(nrrd_head)
#         transform = transforms.ToTensor()
#         voxel = transform(nrrd_data)
#         voxel = voxel.view(1, 32, 32, 32)
    
        pcd_path = os.path.join(data_root, subfolder, 'model.pcd')
        with open(pcd_path, mode='r') as pcd_data:
            # skip first 9 lines
            for i in range(9):
                pcd_data.readline()
            num_line = pcd_data.readline()
            num_points = num_line.split(' ')[1]
            pcd_data.readline()  # skip one more line
            points = []
            print(num_points)

            for i in range(int(num_points)):
                line = pcd_data.readline()
                point = line.split(' ')
                points.append([float(point[0]),
                               float(point[1]),
                               float(point[2])])

        points = torch.tensor(points)
        max, _ = torch.max(points, dim=0)
        min, _ = torch.min(points, dim=0)
        points = points + 0.5 #以0，0，0为原点，边长为1的正方体包围盒
        points = points * 32
        points = random.sample(points,1000)
        
        # update point cloud data
        update_pcd_path = os.path.join(data_root, subfolder, 'model_uniform.pcd')
        with open(update_pcd_path, mode='w') as pcd_data:
            for i in range(len(points)):
                pcd_data.write('{} {} {}\n'.format(float(points[i][0]),
                                                   float(points[i][1]),
                                                   float(points[i][2])))
```

```python
data_root = '..\\data\\bench'
mark = '-skip'
file_index = '4'
for foldName, subfolders, filenames in os.walk(data_root):     #用os.walk方法取得path路径下的文件夹路径，子文件夹名，所有文件名
    for subfolder in subfolders:     #遍历列表下的子文件夹名
#         print(subfolder)
        if subfolder.find(mark) != -1:
            continue
        
        if file_index is not subfolder:
            continue
        
        # read nrrd voxel data
        nrrd_path = os.path.join(data_root, subfolder, 'model.nrrd')
        nrrd_data, nrrd_head = nrrd.read(nrrd_path)
        # print(nrrd_head)
        transform = transforms.ToTensor()
        voxel = transform(nrrd_data)
        voxel = voxel.view(1, 32, 32, 32)

        
        # update point cloud data
        points=[]
        update_pcd_path = os.path.join(data_root, subfolder, 'model_uniform.pcd')
        with open(update_pcd_path, mode='r') as pcd_data:
            for i in pcd_data:
                xyz=i.strip('\n').split(' ')
                points.append([float(xyz[0]),float(xyz[1]),float(xyz[2])])
        points = torch.tensor(points)
        break
                
```

```python
%matplotlib qt
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='r', s=1)
ax.view_init(45, 45)
ax.voxels(voxel[0], color='w', edgecolor="b")
plt.show()
```
